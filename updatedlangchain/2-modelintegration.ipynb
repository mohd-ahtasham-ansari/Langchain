### Streaming and Batch

**Streaming:** Imagine you are talking to someone on the phone. As they speak, you hear their words one by one. Streaming in LangChain is just like that! Instead of waiting for the AI to finish its entire answer before showing it to you, it sends you the answer word-by-word as it's being generated. This makes the AI feel much faster and more responsive.

**Batch:** Imagine you have a stack of 10 different letters you need to mail. Instead of going to the post office 10 separate times, you take them all at once. Batch in LangChain works the same way! If you have multiple questions or tasks for the AI, you can send them all together in one go. The AI processes them all and gives you all the answers back together. This saves a lot of time when you have many things to process.

### stream

```python
for chunk in model.stream("write a paragraph on agentic ai and all tools needed in it n 200 words"):
    print(chunk.text, end="", flush=True)
```

### batch


```python
responses=model.batch([
    "what is ai",
    "what are transformers and agentic ai",
    "what are diifference between ml and ai in term of tech stack used"
])
for response in responses:
    print(response)
```

```python
## setting concurrency request limit
responses=model.batch([
    "what is ai",
    "what are transformers and agentic ai",
    "what are diifference between ml and ai in term of tech stack used"],
    config={
        'max_concurrency':5, #limit is 5 maximum calls
    }
)
for response in responses:
    print(response)
```
